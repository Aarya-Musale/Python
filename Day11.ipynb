```python
# while loop: conditionally infinite loop

# Syntax: while <condition>:
                # until this condition is True loop will RUN
```

```python
num = 10
while num > 5:
  print("Hello",num)
  num -= 1
```

```python
# Sum of first 10 natural numbers
# 1-10 numbers

num = 1
total = 0

while num < 11:
  #print(num)
   total += num
   num += 1
print('Total sum is',total)
```

```python
# while to check the correct pin

#pin = input("Enter your pin:")
import getpass
pin = getpass.getpass("Enter your pin:")

while pin != '1234':
  print("Wrong pin entered")
  pin = getpass.getpass("Please enter correct pin:")
print("Login Successful..!")
```

```python
# Guess numbers using while
secret = 22
guess = 0

while guess != secret:
  guess = int(input("Guess the number:"))
print("Correct Guess...!")
```

```python
#3. Transfer statement
#1. break - exit from current loop
#2. continue - skip current condition and continue the loop
#3. pass - used to create an empty blocks
```

```python
f = ['mango','apple','cherry','apple','grapes','mango','banana']

for item in f:
  if item == 'grapes':
    break
  else:
    print("Process:",item)
```

```python
products = ['mobile','laptop','headphones','mouse','keyboard']
search_item = input("Enter item to search:")

for item in products:
  print("checking",item)

  if item == search_item:
    print("Product found:",item)
    continue
```

```python
# continue : will help to continue the loop and skip current condition or item

products = ['mobile','laptop','headphone','mouse','keyboard','mobile','laptop','headphone']

search_item = input('Enter item to search:')

for item in products:
  print('Checking:',item)

  if item == search_item:
    print('Product found:',item)
    continue
```

```python
marks = [85,90,76,0,70,68,-1,88,96]

for mark in marks:
  if mark == -1:
    print("Invalid marks found,Stop checking")
    break

  if mark == 0:
    print("Absent Student,Skip this student")
    continue

  if mark < 40:
    pass

  print("Valid marks:",mark)
```

**List Comprehension**

- to create a new list in simple onliner solution
- it uses the ternary operator to check if else option

```python
# Syntax: [operation for var in sequence/]
```

```python
d = [2,3,4,5,6,7]
# create a new list with squared numbers
[num * num for num in d]
```

```python
names = ['prasad','imran','gayatri','aarya','bhavana','jay']
[nm.title() for nm in names]
```

```python
[(nm.title(),len(nm)) for nm in names]
```

```python
# Conditional list comprehension
# Fetch those names whose len is> 6
# Syntax: operation for var in sequence/iteration if <condition>
# first element fetched in nm---> chech condition len(nm) > 6---> perform operation

[nm.title() for nm in names if len(nm)>6]
```
