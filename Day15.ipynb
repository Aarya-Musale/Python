```python
# Que: Can you write syntax of lambda function\

#Squaring of number
sq = lambda num: num * num
sq(8)
```

```python
# LAmbda function by default returns the value of the expression

change = lambda name: name.upper()
result = change('aarya')
print(result)
```

```python
# Use of ternary operator in lambda

test = lambda num: '+ve' if num>0 else '-ve'
test(90)
```

```python
# check if number is even or odd

check = lambda num: 'even' if num % 2 == 0 else None
check(8)
```

**Higher Order Function**

```python
# Higher order function: It is a function who need another function as an input
# function(function)
```

```python
# map,filter and reduce are higher order functions
```

```python
# iterable- collection of elements---> str,list,tuple,set,dict,range,....
# syntax:map(function,iterable)

k = [20,5,4,60,10]

# apply an operation over each element from an iterable

map(lambda num: num*num,k)
```

```python
# Rule -When we are gettinf the obj an output in the form of on=bj then we have 2 choices
#choice1 = typecaste the obj to  see actual values

set(map(lambda num: num*num,k))
```

```python
# choice2 - use for loop

out = map(lambda num: num*num,k)
for val in out:
  print(val)
```

```python
# can we use multiple args in lambda:YES

test = lambda a,b,c: a+b
test(100,800,500)
```

```python
test = lambda a,b,c: max(a,b,c)
test(100,800,1500)
```

```python
names =('gayatri','abdul','snehal','jay')

map(lambda upper_case:upper_case.upper(),names)
```

```python
list(map(lambda upper_case:upper_case.upper(),names))
```

```python
operation = lambda names:names.upper()
list(map(operation,names))
```
