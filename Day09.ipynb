```python
#1. Selective statements: these statements are conditional based
#we use conditional operators to check or compare the objects
#Conditional operatores: < > <= >= == != these operators generate boolean answers

#In order to use these statements/blocks we have if elif else
```

```python
#When user wants to test single condition then use only if <condition>
```

```python
name  = 'aarya'
if name == 'aarya':
  print('Good Morning')
```

```python
#if condition result is True then ur program control will enter inside the block
```

```python
#if we want to check 2 conditions in one if block the use operators (and, or, not)

age = 23 #int(input('Enter your age:'))
country = 'india'

if age >= 18 and country == 'india':
  print('Eligible to vote')

else: #default when if condition result is False
  print('Not Eligible to vote')
```

```python
#if for one check
#if else for 2 check
#if use want to check more than 2 options then use elif

amt = float(input("Enter your amount:"))

#Whenever user takes a block in python it ends with: and at the next line
#your body of the block starts
#all statements we write inside that block.The space infront of the each
#statement is called Indentation

if amt > 100000:
  print('Purchase iphone')
elif amt > 90000:
  print('Purchase Samsung/Vivo/RealMe phones')
elif amt > 50000:
  print('Purchase Oppo/Lava...etc phones')
else:
  print('You can try with remaining items')
```

```python
#2. Iterative statement
#It contains 2 loops
#1. for loop
#2. while loop
```

```python
#for loop: is an iterative loop which feteches one element at a time from the collection
a  = [10,20,30,40]

#Syntax: for var in collection/iterable

for val in a[::-1]:
  print(val)
```

```python
name = 'aarya'
for i in name:
  print(i)
```

```python
#Take a dict and apply for loop

d1 = {'a':10,'b':20,'c':30,'d':40}
for val in d1:
  print(val)

#Here we are getting only keys
```

```python
#to get keys and values both
d1 = {'a':10,'b':20,'c':30,'d':40}
for val in d1:
  print(val,d1[val])
```

```python
#To get keys separate and values separate
d1
```

```python
for val in d1.items():
  print(val)
```
