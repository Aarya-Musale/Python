```python
# Accept an integer from the user and display it as an octal number (Base 8).

num = int(input("Enter a Number:"))
print(f"Octal number of {num} is {oct(num)}")
```

```python
# Display a float number (a number with decimals) such as 458.541315 rounded to exactly two decimal places.

num = float(input("Enter a Number:"))
print('%.2f'% num)
```

```python
# Ask the user for a numerator and a denominator.
# Calculate the percentage (numerator/denominator * 100) and display it with exactly two decimal places followed by a percent sign.

numerator = float(input("Enter a numerator:"))
denominator = float(input("Enter a denominator:"))

percentage = (numerator/denominator * 100)
print(f'The Percentage is:{percentage:.2f}%')
```

```python
# Ask the user for a number.
# Print this number padded with leading zeros so the total width is exactly 5 digits.

number = int(input("Enter a number:"))
formatted_num = f'{number :05}'
formatted_num
```

```python
# Given three variables quantity = 3, totalMoney = 450, and price = 150, use the format() method to print a sentence that neatly displays these values.

quantity = 3
totalMoney = 450
price = 150

statement = 'I have Rs{1} so I can buy {0} flower-pots for Rs{2:.2f}'

print(statement.format(quantity,totalMoney,price))
```

```python
# Write a program that accepts 5 float numbers as input from the user and stores them in a list.

numbers = []

for i in range(0,5):
  print("Enter number at location",i,':')
  items = float(input())
  numbers.append(items)

print("User List:",numbers)
```

```python
# You have two lists: names = ["Alice", "Bob", "Charlie"] and scores = [85, 92, 78].
# Print these as a table with aligned columns.

names = ["Alice", "Bob", "Charlie"]
scores = [85, 92, 78]

print(f'{'Name':<10}{'Score'}')
print('-' * 15)

for name, score in zip(names,scores):
  print(f"{name:<10} {score}")
```

```python
# Create a menu that offers three options: “1. Say Hello”, “2. Calculate Square”, and “3. Exit”.
# The program should perform the action based on the number the user types.

print('1. Say Hello\n2. Calculate Square\n3. Exit')
choice = int(input("Enter Choice(1-3):"))

if choice == 1:
  print("Hello there! Hope you're having a great day.")
elif choice == 2:
  num = int(input("Enter a number to calculate square:"))
  print(f'Square of {num} is {num*num}')
elif choice == 3:
  print('Exit')
else:
  print('Invalid Input.')
```
