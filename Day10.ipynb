```python
#It is used to fetch the elements from collection ONE by ONE
names = ['vandana','sarita','tauzi','prasad','pratik']
for n in names:
  print(n.upper())
```

```python
names
```

```python
for n in names:
  if n.endswith('a'):
    print(n.upper())
```

```python
# fetch names in which i is present

for n in names:
  if 'i' in n:
    print(n.title())
```

```python
# fetch names whose len is <=5

for n in names:
  if len(n) <= 5:
    print(n)
```
