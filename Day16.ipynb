```python
# function(function)--->Higher order function
# map,filter,reduce

# map(function,iterable)
# filter(function,iterable)

# Que: map vs filter?
```

```python
# reduce(function,iterable)---> used to convert collection into a single value/obj
# reduce is present in functools module


# For example, reduce(lambda x,y:x+y,)

from functools import reduce

k =[2,3,4,5,6]

reduce(lambda x,y:x+y,k) # in reduce we do not need typecasting as it gives the result
```

```python
reduce(lambda x,y:max(x,y),[0,120,52,125])
```

**Modules in Python**

```python
# a file with .py extension is a module
# means any python file is a module

# module contains = varibales + functions + class + constant value CONST + methods
# So user/developer can use the members of a module
```

```python
# We have 2 types of modules
#1. Built-in module ---> modules present in python itself
#2. USer defines module ---> made by user
```

```python
#1. Built-in module ---> modules present in python itself
# syntax: import module_name

import math
```

```python
math.e  # Eulars constant
```

```python
math.pi
```

```python
math.factorial(5)
```

```python
# module aliasing ---> give a short name or nick name
# it is required when we have bigger module names then we want

import math as m
m.sqrt(625)
```

```python
m.pow(2,3)
```

```python
# to import only specific and required things we need to go with
# from module_name import member
```
