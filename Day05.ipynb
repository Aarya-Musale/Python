```python
#Tuple packing and unpacking
s = 100
u = 'aarya'
v = 300
```

```python
#Packing: Combine segregated/seperated elements as a one entity
s,u,v
```

```python
chars = s,u,v
chars
```

```python
#Unpacking: Process of separating an individual object and store ot in separate variable is called Tuple Unpacking
chars
```

```python
char1,char2,char3 = chars
char1
```
