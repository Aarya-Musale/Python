**Lambda Function**

```python
# Write a lambda function that takes one argument and returns its square.

square = lambda num:num*num
square(25)
```

```python
# Create a lambda function to check if a number is even (returns True or False).

check = lambda num:True if num % 2 == 0 else False
print(check(25))
print(check(22))
```

```python
# Use a lambda function to add two numbers together.

add = lambda a,b:a+b
add(20,16)
```

```python
# Write a lambda function that takes a string and returns it in uppercase.

word = lambda str1:str1.upper()
word('python')
```

```python
# Create a lambda function that calculates the cube of a number.

cube = lambda num: num*num*num
cube(3)
```

```python
# Use a lambda function to return the length of a string.

length = lambda str1: len(str1)
length('science')
```

```python
# Write a lambda function that takes a number and returns its absolute value.

abs_val = lambda x:x if x>=0 else -x
abs_val(-6)
```

```python
# Create a lambda function to find the maximum of two numbers.

max_num = lambda a,b:max(a,b)
max_num(10,52)
```

```python
# Use a lambda function to check if a string starts with the letter 'A'.

check = lambda str1: True if str1.startswith('A') else False
check('Aarya')
```

```python
# Write a lambda function that returns the last element of a list.

last_ele = lambda ele:ele[-1] if ele else None

last_ele([10,20,30])
```

**Map function**

```python
# Given nums = [1, 2, 3, 4], use map() and a lambda to return a list of squares.

nums = [1, 2, 3, 4]
list(map(lambda num:num*num,nums))
```

```python
# Given words = ['hello', 'world'], use map() to convert all strings to uppercase.

words = ['hello', 'world']
list(map(lambda word:word.upper(),words))
```

```python
# Given nums = [10, 20, 30], use map() to add 5 to each element.

nums = [10, 20, 30]
list(map(lambda num:num+5,nums))
```

```python
# Given temps_c = [0, 20, 37], use map() to convert them to Fahrenheit ($F = C \times 9/5 + 32$).

temps_c = [0, 20, 37]
list(map(lambda temp:(temp * 9 / 5) + 32,temps_c))
```

```python
# Given strings = ['1', '2', '3'], use map() to convert them to integers.

strings = ['1', '2', '3']
list(map(lambda l1:int(l1),strings))
```

```python
# Given nums = [1, 2, 3], use map() to return the list of floats (e.g., [1.0, 2.0, 3.0]).

nums = [1, 2, 3]
list(map(lambda l1:float(l1),nums))
```

```python
# Given a list of strings, use map() to return the length of each string.

string_list = ['data','science','machine','learning']
list(map(lambda length:len(length),string_list ))
```

```python
# Given nums = [2, 4, 6], use map() to return the cube of each number.

nums = [2, 4, 6]
list(map(lambda num:num*num*num,nums))
```

```python
# Given names = ['alice', 'bob'], use map() to capitalize the first letter of each name.

names = ['alice', 'bob']
list(map(lambda x:x.capitalize(),names))
```

```python
# Given nums = [10, 20, 30], use map() to return the boolean result of x > 15 for each element.

nums = [10, 20, 30]
list(map(lambda x:True if x>15 else False,nums))
```

**Filter function**

```python
# Given nums = [1, 2, 3, 4, 5, 6], use filter() to keep only the even numbers.

nums = [1, 2, 3, 4, 5, 6]
list(filter(lambda x:x if x % 2 == 0 else None,nums))
```

```python
# Given nums = [-2, -1, 0, 1, 2], use filter() to keep only positive numbers.

nums = [-2, -1, 0, 1, 2]
list(filter(lambda x:x if x>0 else None,nums))
```

```python
# Given words = ['apple', 'bat', 'car', 'dog'], use filter() to keep words with length greater than 3.

words = ['apple', 'bat', 'car', 'dog']
list(filter(lambda word:word if len(word)>3 else None,words))
```

```python
# Given nums = [10, 15, 20, 25, 30], use filter() to keep numbers divisible by 10.

nums = [10, 15, 20, 25, 30]
list(filter(lambda num:num if num % 10 ==0 else None,nums))
```

```python
# Given chars = ['a', 'b', 'C', 'D'], use filter() to keep only uppercase letters.

chars = ['a', 'b', 'C', 'D']
list(filter(lambda ch:ch if ch.isupper() else None,chars))
```

```python
# Given data = [None, 'hello', '', 'world'], use filter() to remove empty/None values.

data = [None, 'hello', '', 'world']
list(filter(lambda d:d is not None and d != '' ,data))
```

```python
# Given nums = [1, 5, 12, 18, 25], use filter() to keep numbers greater than 15.

nums = [1, 5, 12, 18, 25]
list(filter(lambda num:num if num>15 else None,nums))
```

```python
# Given emails = ['a@b.com', 'invalid', 'c@d.com'], use filter() to keep strings containing '@'.

emails = ['a@b.com', 'invalid', 'c@d.com']
list(filter(lambda l:l if '@' in l else None,emails))
```

```python
# Given nums = [1, 2, 3, 4, 5], use filter() to keep only odd numbers.

nums = [1, 2, 3, 4, 5]
list(filter(lambda num:num if num % 2 != 0 else None,nums))
```

```python
# Given names = ['Anna', 'Bob', 'Alex', 'Charlie'], use filter() to keep names starting with 'A'.

names = ['Anna', 'Bob', 'Alex', 'Charlie']
list(filter(lambda n:n if n.startswith('A') else None,names))
```
