**String**

```python
# String is immutable datatype
# we cannot change the string once it is created and we cant modify or update the string
# String background data structure is an array
# Array is a collection of sequential elements
# s = 'python'
# p---> 0
# y---> 1
# t---> 2
# so on...
```

```python
s = 'PYTHON'

# indexing access only single element at a time
s[-1]
```

```python
# Indexing is of 2 types
# +ve indexing---> starts with 0 and ends at (n-1) n means total chars/block (left-right)
# -ve indexing---> starts with -1 and ends at -n (right-left)
```

```python
# Slicing [start:stop:step]
# using this we can access more than one element
```

```python
s = 'PYTHON'
s[-3:-1]  #[start:stop] stop is exclusive means elements at this index position will not be in the output
```

```python
# Assignment
# Access PY
# Access YTH
# Access HO
```

```python
s = 'PYTHON'
s[:2],s[1:4],s[3:5]
```
