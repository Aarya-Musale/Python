**Function with return keyword**

```python
#What is the advantage of function
# Function will work on specific requirement
# this result we need to use as an input to the outside of the function
# for this purpose we need to use return keyword

def test(balance):
  if balance <= 0:
    return "No"
  else:
    return "Yes"

test(120)
```

```python
def test(balance):
  if balance <= 0:
    return "No"
  else:
    return "Yes"

result = test(120)

if result == 'Yes':
  print("You can perform transactions")
else:
  print("You cannot perform transactions")

#Hence, when we want to bring the result of a function from inside to outside then
# we need return to be used
```

```python
# Purchase and billing example using return

items = {'rice':56,'chana':30,'wheat':60,'milk':35,'parle_g':35}
bill = []

def purchase(*args):
  for item in args:
    #print(item,items[item])
    bill.append(items[item])
  return bill

def check(total):
  return sum(total)

#calling of purchase
total = purchase('wheat','rice','parle_g')
print("item amounts:",total)

#calling of check
ans = check(total)
print("Final bill amount is Rs",ans)
```

```python
# Calculate % and calaculate grade

def calculate_per(marks_obtained,total_marks):
  return (marks_obtained / total_marks)*100
#print(calculate_per(450,500))


def calculate_grade(per):
  if per >= 75:
    return 'Distition'

  elif per >= 65:
    return 'First Class'

  elif per >= 55:
    return 'Second Class'

  elif per >= 45:
    return 'Pass Class'

  else:
    return "Fail"

per  = calculate_per(440,500)
result = calculate_grade(per)

print("Total Percentage:",per)
print("Grade:",result)
```
