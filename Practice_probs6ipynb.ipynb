```python
# Python Program to Swap Two Variables
x = int(input("Enter a 1st number:"))
y = int(input("Enter a 2nd number:"))

temp = x
x = y
y = temp
print(f"Value of x after swapping:{x}")
print(f"Value of y after swapping:{y}")
```

```python
# Python Program to Check if a Number is Positive, Negative or 0
num = int(input("Enter a number:"))
if num > 0:
  print(f"{num} is Positive!")
elif num < 0:
  print(f"{num} is Negative!")
else:
  print("Number is zero.")
```

```python
# Python Program to Check if a Number is Odd or Even
num = int(input("Enter a number:"))
if num % 2 == 0:
  print(f"{num} is Even!")
else:
  print(f"{num} is Odd!")
```

```python
# Python Program to Check Leap Year
year = int(input("Enter Year to Check:"))

if (year % 4 == 0) and (year % 100 != 0):
  print(f"{year} is a Leap Year!")
else:
  print(f"{year} is not a Leap Year!")
```

```python
# Python Program to Find the Largest Among Three Numbers
num1 = int(input("Enter 1st Number:"))
num2 = int(input("Enter 2nd Number:"))
num3 = int(input("Enter 3rd Number:"))
if (num1 > num2) and (num1 > num3):
  print("num1 is the Greatest!")
elif (num2 > num3) and (num2 > num1):
  print("num2 is the Greatest!")
else:
  print("num3 is the Greatest!")
```

```python
# Python Program to Print all Prime Numbers in an Interval
nums = [2,3,5,7,10,11,12,13,15,19]
primes = [num for num in nums
          if num >1 and all(num % i != 0 for i in range(2,num))]
primes
```

```python
# Python Program to Find the Factorial of a Number
num = int(input("Enter a number:"))
factorial = 1

if num < 0:
  print("Sorry!!,Factorial does not exist for negavtive numbers.")
elif num == 0:
  print("Factorial of 0 is 1.")
else:
  for i in range(1,num + 1):
    factorial *= i
  print(f"Factorial of {num} is:{factorial}")
```
