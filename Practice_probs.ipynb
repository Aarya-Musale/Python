```python
# Write a Python program to print all natural numbers from 1 to n.

n = int(input("Enter a number:"))
print(f"Natural numbers from 1 to {n} are:")
for i in range(1,n+1):
  print(i,end = ' ')
```

```python
# Write a Python program to print all natural numbers in reverse (from n to 1)

n = int(input("Enter a number:"))
print(f"Natural numbers from {n} to 1 in reverse are:")
for i in range(n, 0, -1):
  print(i,end = ' ')
```

```python
# Write a Python program to print multiplication table of any number. Take user input.

n = int(input("Enter a number:"))
print(f"Multiplication table for {n}:")
for i in range(1,11):
  result = n * i
  print(f"{n} x {i} = {result}")
```

```python
# Write a Python program to calculate sum of digits of a number.

n = input("Enter a number:")
sum = 0
for char in n:
  if char.isdigit():
    sum += int(char)

print(f"Sum of digits of is = {sum}")
```

```python
# Write a Python program to calculate product of digits of a number.

n = input("Enter a number:")
product = 1
for char in n:
  if char.isdigit():
    product *= int(char)

print(f"Product of digits is= {product}")
```
