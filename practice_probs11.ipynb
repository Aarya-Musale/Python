```python
# Add Two Numbers Using Lambda

add = lambda x,y: x+y
print('Addition of two numbers is:',add(4,23))
```

```python
# Find Square and Cube Using Lambda

sq = lambda num1:num1*num1
print('square of a number is:',sq(5))

cube = lambda num2:num2 ** 3
print('cube of a number is:',cube(4))
```

```python
# Check Positive, Negative, or Zero

test = lambda num:'+ve' if  num > 0 else ('-ve' if num<0 else 'zero')
print(test(12))
print(test(-9))
print(test(0))
```

```python
# Filter Words Starting with a Specific Letter

words = ["Apple", "Banana", "Avocado", "Mango", "Apricot"]
result = list(filter(lambda x:x.startswith('A'),words))
print('Words starting with A are:',result)
```

```python
# Convert List of Strings to Uppercase

str1 = ['black','yellow','orange','pink']
result = list(map(lambda x:x.upper(),str1))
print(result)
```

```python
# Sort Students by Marks in Descending Order

students = [('Anshita',96),('Raj',85),('Anvita',95),('Harsh',88)]

result = sorted(students,key=lambda x:x[1],reverse = True)
result
```

```python
# Sort Employees by Name Length

employees = ['Pratik','Neha','Rahul','Jay','Riya']

result = sorted(employees,key=lambda name:len(name))
result
```

```python
# Find Maximum Salary Employee

employee = [
          {'name':'Snehal','salary':60000},
          {'name':'Riya','salary':72000},
          {'name':'Harsh','salary':45000}
          ]

result = max(employee,key=lambda emp:emp['salary'])
result
```

```python
# Remove Empty Strings from a List

lst1 = ['hello','','world','','data science','']
result= list(filter(lambda x:x != '',lst1))
result
```

```python
#  Calculate Final Price After Discount

price = lambda p,dis:p-(p*dis/100)
price(1500,10)
```

```python
#  Use Lambda with reduce() to Find Maximum

from functools import reduce
max_num = [10,54,22,48,14]
result = reduce(lambda num1,num2:num1 if num1>num2 else num2,max_num)
result
```

```python
#  Lambda vs def Function

def check_def(x):
  return x * 2

check_lambda = lambda x:x * 2

print(check_def(4))
print(check_lambda(4))
```
