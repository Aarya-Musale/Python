```python
#  Write a Python program to create a function that can add any number of values.

def add_nums(*nums):
  total = 0
  for n in nums:
    total += n
  return total

add_nums(10,20,40)
```

```python
# Create a function that accepts three numbers and returns the largest number.

def largest_num(x,y,z):
  if (x > y) and (x > z):
    return x

  elif (y > x) and (y > z):
    return y

  else:
    return z

print('Largest Number is:',largest_num(10,52,85))
```

```python
#  Create a function that returns the factorial of a given number.

def factorial(n):
  if n == 0:
    return 1
  return n * factorial(n-1)

print('Factorial is:',factorial(4))
```

```python
#  Create a function that checks whether a number is prime or not.

def is_prime(num):
  if num <= 1:
     return False

  i = 2
  while num < i:
    if num % i == 0:
      return False
    i += 1
  return True

print(is_prime(19))
```

```python
#  Create a function that accepts a string and returns the reversed string

def reverse(str1):
  return str1[::-1]

reverse('Python')
```

```python
# Create a function that counts vowels in a given string.

def vowels_count(str1):
  vowels = 'aeiouAeiou'
  count = 0

  for ch in str1:
    if ch in vowels:
      count += 1
  return count

vowels_count('Data Science')
```

```python
#  Create a function that checks whether a string is palindrome or not.

def is_palindrome(text):
  cleaned = text.lower()
  return cleaned == cleaned[::-1]

print(is_palindrome('Madam'))
```

```python
#  Create a function that removes duplicate values from a list while maintaining order.

def remove_duplicates(items):
  result = []

  for item in items:
    if item not in result:
      result.append(item)

  return result

print(remove_duplicates([1,5,2,4,6,1,3,4,5]))
```
