```python
#  Write a Python program to square every number in a list using map().

lst = [2,3,4,5]
result  = list(map(lambda x:x**2,lst))
result
```

```python
#  Convert a list of Celsius temperatures into Fahrenheit using map().

lst = [12,30,24,50]
result = list(map(lambda c:(c * 9/5)+32,lst))
result
```

```python
#  Given a list of student dictionaries, extract only the names using map().

students = [
    {'name': 'Amit', 'marks': 80},
    {'name': 'Pranav', 'marks': 92},
    {'name': 'Rahul', 'marks': 76}
]

result = list(map(lambda n:n['name'],students))
result
```

```python
# Write a Python program to filter only positive numbers from a list.

nums = [-5, 10, -2, 0, 7, 3]
result  = list(filter(lambda x:x>0,nums))
result
```

```python
#  Filter students whose marks are greater than or equal to 40.

students = [
    {'name': 'Amit', 'marks': 35},
    {'name': 'Pranav', 'marks': 68},
    {'name': 'Rahul', 'marks': 40}
]

result = list(filter(lambda x:x['marks']>=40,students))
result
```

```python
#  Filter all words that start with the letter P.

words = ['Python', 'Java', 'Pandas', 'AI', 'PowerBI']

result = list(filter(lambda w:w.startswith('P'),words))
result
```

```python
#  Write a Python program to find the sum of list elements using reduce().

from functools import reduce
numbers = [5, 10, 15, 20]
result = reduce(lambda a,b:a+b,numbers)
result
```

```python
#  Find the smallest number from a list using reduce().

from functools import reduce
numbers = [45, 18, 72, 9, 30]
result = reduce(lambda x,y:x if x<y else y,numbers)
result
```

```python
#  Find the total number of characters in a list of words using map() and reduce().

from functools import reduce
words = ['AI', 'Python', 'Data']
len_words = list(map(lambda x:len(x),words))
total_chrs = reduce(lambda x,y:x+y,len_words)
total_chrs
```

```python
#  Use map() and reduce() to apply a 10% discount to prices and calculate the final total bill.

from functools import reduce
prices = [1000, 500, 800]
discount = list(map(lambda x:x - (x * 10/100),prices))
bill = reduce(lambda x,y:x+y,discount)
bill
```

```python
# From a list of numbers, filter even numbers and find their sum using filter() and reduce().

from functools import reduce
numbers = [1, 2, 3, 4, 5, 6]
even_nums = list(filter(lambda x:x %2 == 0,numbers))
total = reduce(lambda x,y:x+y,even_nums)
total
```

```python
# Use lambda with sorted() to sort employees based on salary.

employees = [
    ('Amit', 45000),
    ('Pranav', 60000),
    ('Rahul', 35000)
]

result = sorted(employees,key=lambda x:x[1])
result
```
