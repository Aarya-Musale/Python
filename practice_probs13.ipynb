```python
# Use map() to create a new list containing the square of each number.

lst = [1,2,3,4,5]

result = list(map(lambda x:x ** 2,lst))
result
```

```python
# Convert a list of numeric strings into integers using map()

lst = ['10','32','53','24','65']

result = list(map(lambda x:int(x),lst))
result
```

```python
# Add corresponding elements of two lists using map()

lst1 = [1,5,6]
lst2 = [12,14,15]

result = list(map(lambda a,b:a+b,lst1,lst2))
result
```

```python
# Use filter() to select only even numbers from a list

lst = [12,21,52,23,14,15]
result = list(filter(lambda n:n %2 == 0,lst))
result
```

```python
# Remove empty strings from a list using filter().

lst = ['AI','','Machine Learning','','Hello World']
result = list(filter(lambda x:x != '',lst))
result
```

```python
# Filter names whose length is greater than 5 characters

names = ['Aarya','Anshita','Shivani','Swati','Shantilal']
result = list(filter(lambda x:len(x)>5,names))
result
```

```python
# Use reduce() to find the sum of all numbers in a list.

from functools import reduce
lst = [12,14,15]
result = reduce(lambda x,y:x+y,lst)
result
```

```python
#  Use reduce() to multiply all numbers in a list

from functools import reduce
lst = [2,4,5]
result = reduce(lambda x,y:x*y,lst)
result
```

```python
# Find the largest number from a list using reduce()

from functools import reduce
lst = [28,44,35]
result = reduce(lambda x,y:x if x>y else y,lst)
result
```

```python
# Select even numbers from a list and then square them.

lst = [1,2,3,4,5,6]

even_sq = list(map(lambda x:x * x,filter(lambda x: x %2 == 0,lst)))
even_sq
```

```python
# Apply a normal function to every element of a list using map()

def add(a):
  return a + 25

lst = [2,3,4]
result  = list(map(add,lst))
result
```

```python
#  Use reduce() with an initial value to add numbers.

from functools import reduce
nums = [10,40,20]
result = reduce(lambda x,y:x++y,nums,100)
result
```
