```python
# Write a program to check if two strings are balanced.
#For example, strings s1 and s2 are balanced if all the characters in s1 are present in s2.
# The character’s position doesn’t matter.


def balance_str(s1,s2):
  flag = True
  for char in s1:
    if char in s2:
      continue
    else:
      flag = False
      break
  return flag

s1 = 'Data Science'
s2 = 'Data Science'

print('Is string balanced:',balance_str(s1,s2))
```

```python
# Write a program to count the total number of vowels (a, e, i, o, u) in a given string.

str1 = "Hello World"
vowels = 'aeiouAEIOU'
count = 0

for char in str1:
  if char in vowels:
    count += 1

print("Vowels in the string are",count)
```

```python
#  Check if a given URL starts with “https” and ends with “.com”.

url = 'https://www.NewAirways.com'

is_https = url.startswith('https')
is_com = url.endswith('.com')

if is_https and is_com:
  print("Is valid URL:True")
else:
  print("Is not a valid URL:False")
```

```python
# Write a program to arrange string characters such that all lowercase letters come first, followed by all uppercase letters.

str1 = 'PytHoN'
lower_case = []
upper_case = []

for char in str1:
  if char.islower():
    lower_case.append(char)
  else:
    upper_case.append(char)

updated_str = ''.join(lower_case + upper_case)
print('Result:',updated_str)
```
