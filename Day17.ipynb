**Exception handling**

```python
# Exception is nothing but the error

try:
  #print(30+'A')  #line with error
  print(30+20)    #line without error
except:
  print("Error Occured")


for i in range(0,5):
  print(i)
```

```python
# to handle the exception we must need 2 blocks
# try: to test the code
# except: to handle the exception
# to understand the error msg we need to use exception class

try:
  #print(30+'A')  #line with error
  #print(30+20)   #line without erro
  #print(30/0)
  print(int('ten'))
except Exception as msg:
  print("Error Occured",msg)
```

```python
# Named Exception
try:
  #print(30+'A')  #line with error
  #print(30+20)   #line without erro
  #print(30/0)
  print(int('ten'))
except TypeError as msg:
  print("Error Occured",msg)

except ValueError as msg:
  print("Error Occured",msg)
```

> try:test
> except:to handle exception
> else:if no exception then come here
> finally:irrespective of exception this block will get executed

```python
try:
  #print(10/4) #without exception
  print(10/0) #with exception ZeroDivisionError

  print(int('ten')) # Exception ValueError

except Exception as msg:
  print(msg)
else:
  print('No exception do below tasks...')
finally:
  print('Zukega nai sala....')
```

```python
# Example of ATM operation

try:
  balance = 5000
  amount = int(input('Enter amout in Rs.'))

  if  amount > balance:
    raise ValueError('Insufficient balance..')

except ValueError as e:
  print('Error:',e)

else:
  balance = balance - amount
  print('Withdrawal succesfull....')
  print('Remaining Amount is Rs. ',balance)

finally:
  print('Thank you for banking with us Visit Again.. :-)')
```
