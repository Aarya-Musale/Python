```python
# Write a Python program to print numbers from 1 to 5 using a while loop.

i = 1
while i <= 5:
  print(i)
  i = i + 1
```

```python
# Write a Python program to print numbers from 10 down to 1 using a while loop

i = 10
while i >= 1:
    print(i)
    i = i - 1
```

```python
#  Write a Python program to calculate the sum of numbers from 1 to 10.

i = 1
total = 0

while i <= 10:
  total += i
  i = i + 1

print("Sum is:",total)
```

```python
# Write a python program to print the multiplication table of 5.

num = 5
i = 1

while i <= 10:
  print(num, 'x', i, '=',num * i)
  i = i + 1
```

```python
# Write a Python program to print even numbers from 2 to 20.

i = 2
while i <= 20:
  print(i)
  i = i + 2
```

```python
# Write a Python program to find the factorial of 5 using a while loop.

num = 5
fact = 1

while num > 0:
  fact = fact * num
  num = num - 1

print("Factorial is:",fact)
```

```python
#  Write a menu-driven Python program that displays choices repeatedly until the user selects Exit.

choice = 0

while choice != 4:
  print("1. Add")
  print("2. Substract")
  print("3. Multiply")
  print("4. Exit")

  choice = int(input("Enter your choice:"))

  if choice == 1:
    print("You Selected Add.")
  elif choice == 2:
    print("You Selected Substract.")
  elif choice == 3:
    print("You Selected Multiply.")
  elif choice == 4:
    print("Thank You!")
  else:
    print("Invalid choice!")
```

```python
# Write a Python program that allows only 3 attempts to enter the correct ATM PIN.

pin = 1234
attempts = 3

while attempts > 0:
  enter_pin = int(input("Enter Pin:"))

  if enter_pin == pin:
    print("Login Successful")
    break

  else:
    attempts = attempts - 1
    print("Wrong PIN. Attempts left:",attempts)

if attempts == 0:
  print("Card Blocked")
```

```python
# Write a Python program that repeatedly asks for a name and stops only when the user types exit.

while True:
  name = input("Enter your name or type exit:")

  if name == 'exit':
    break

  print("Hello",name)
```

```python
# Write a Python program to keep accepting marks and calculate the total. Stop when the user enters 0.

total = 0
marks = int(input("Enter your marks:"))

while marks != 0:
  total += marks
  marks = int(input("Enter your marks:"))

print("Total Marks:",total)
```
