```python
#  Create a simple lambda function that adds 10 to a given number.

add = lambda x:x+10
add(12)
```

```python
# Create a lambda function to multiply two numbers

mul = lambda x,y:x * y
mul(2,6)
```

```python
# Create a lambda function that returns the square of a given number

sq = lambda x:x ** 2
sq(6)
```

```python
#  Create a lambda function to check whether a number is even or odd.

even_odd = lambda x:'Even' if x %2 == 0 else 'Odd'
print(even_odd(8))
print(even_odd(1))
```

```python
# Use lambda with map() to square every number in a list

lst = [2,6,4,7,12,5]
result = list(map(lambda x:x ** 2,lst))
result
```

```python
#  Use lambda with filter() to get only even numbers from a list.

lst = [2,6,4,7,12,5]

result = list(filter(lambda n:n %2 == 0,lst))
result
```

```python
# Sort a list of tuples based on the second value using lambda.

data = [('Neha',85),('Amit',92),('Ravi',78)]
result = sorted(data,key=lambda x:x[1])
result
```

```python
#  Sort employee records based on salary using lambda.

employees = [
    {"name": "Amit", "salary": 45000},
    {"name": "Neha", "salary": 60000},
    {"name": "Ravi", "salary": 38000}
]

result = sorted(employees,key=lambda x:x['salary'])
result
```

```python
# Create a lambda function that returns the maximum of two numbers

max_num = lambda x,y:max(x,y)
max_num(12,18)
```

```python
# Create a function that returns a lambda function for multiplying a number by a given value.

def multiplier(n):
  return lambda x: x * n

mul1 = multiplier(2)
mul2 = multiplier(3)

print(mul1(11))
print(mul2(11))
```

```python
#  Use lambda with reduce() to find the product of all numbers in a list

from functools import reduce

lst = [2,7,6,8,4]
result = reduce(lambda x,y:x*y,lst)
result
```

```python
#  Write the same logic using a normal function and a lambda function.

def add(a,b):
  return a+b

add_lambda = lambda a,b:a+b

print(add(10,15))
print(add_lambda(10,15))
```
