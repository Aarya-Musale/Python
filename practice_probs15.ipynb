```python
#  Handle division by zero so that the program does not crash.

a = 10
b = 0
try:
  result = a/b
  print(result)
except Exception as e:
  print('Error Occured:',e)
```

```python
# Use try-except to handle invalid numeric input.

try:
  age = int(input("Enter your age:"))
  print("Your age is:",age)
except Exception:
  print("Please Enter Valid Age..")
```

```python
#  Handle both invalid input and division by zero in one program.

try:
  a = int(input("Enter 1st Number:"))
  b = int(input("Enter 2nd Number:"))
  print('Result:',a/b)
except ValueError:
  print("Invalid Input.Please Enter numbers only")

except ZeroDivisionError:
  print("Cannot Divide by Zero.")
```

```python
# Catch an exception and display the actual error message.

try:
  numbers = [10,20,30]
  print(numbers[5])
except Exception as e:
  print("Error Occured:",e)
```

```python
# Use else to run code only when no exception occurs

try:
  num = int(input("Enter a number:"))
except ValueError:
  print("Invalid input")

else:
  print('You Entered',num)
```

```python
# Use finally to execute code whether an exception occurs or not.

try:
  pin = int(input("Enter Your Pin:"))
except ValueError:
  print("Wrong Pin Entered.Please enter correct Pin...")

finally:
  print("Thank You!!")
```

```python
# Raise an exception if age is less than 18.

age = int(input("Enter your age:"))

try:
  if age < 18:
    raise ValueError('Age must be 18 or above')
  print("Eligible")
except ValueError as e:
  print(e)
```

```python
# Create a custom exception for insufficient balance.

class InsufficientBalanceError(Exception):
    pass

balance = 500
withdraw = 1000

try:
  if withdraw > 500:
    raise InsufficientBalanceError('Insufficient account balance')
  balance = balance - withdraw
  print('Withdrawal Successful')
except InsufficientBalanceError as e:
  print(e)
```

```python
#  Safely read a file and handle the situation when the file does not exist.

try:
    with open("data.txt", "r") as file:
        content = file.read()
        print(content)
except FileNotFoundError:
    print("The file does not exist")
```

```python
#  Use nested try-except to handle errors at different levels.

try:
    num = int(input("Enter a number: "))
    try:
        result = 100 / num
        print("Result:", result)
    except ZeroDivisionError:
        print("Cannot divide by zero")
except ValueError:
    print("Invalid input")
```

```python
#  Handle an error when trying to access an invalid list index.

items = ["Python", "Java", "SQL"]
try:
    print(items[4])
except IndexError:
    print("Invalid index selected")
```

```python
# Create a function that safely divides two numbers.

def safe_divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return "Cannot divide by zero"
print(safe_divide(10, 2))
print(safe_divide(10, 0))
```
