**Fuction with Arguements**

```python
def addition(x,y):
  print('Addition:',x+y)
  print('Multiplication:',x*y)
  print('Substraction:',x-y)
  print('division:',x/y)

addition(100,200)
```

**Types of Arguements**
- Positional Arguement
- Keyword Arguement
- Default Arguement
- Variable Length Arguement

```python
# Positional Arguemnts: these are args. in which sequence order matters

def test(x,y,z):
  print(x,y,z)

test(100,200,300)
test(200,100,300)
```

```python
def info(name,age,sal):
  print('Your name is',name,'your age is',age, '& your salary is',sal  )

info('Vicky',23,50000) #Positional args
```

```python
info(60000,'Prasad',23) # If user does not follow order then output is not as per the expectation
```

```python
# To solve above problem we have keyword arguement
# It does not require to follow order
# Bcz it directly refers to key=value format

info(sal = 60000,name = 'Anshita',age = 23)
```

```python
info("Vicky",22) #we must supply 3 values to 3 args
```

```python
info(name ='Imran') # in keyword also we need to supply the values = number of args
```

```python
# To solve above problem we have Default Argement
# using default args we set the static values in declaration
# So when user is not supplying the values from calling

def info(name='Guest',age=25,sal=35000):
  print('Your name is',name,'your age is',age, '& your salary is',sal  )

info()
```

```python
info('Prasad')
```

```python
# when user is sending the new values then those will be taken
# else it will take default values
```

```python
info(25,'prasad')
```

```python
info(age = 25,name = 'prasad')
```

```python
# Rule - we can use positional and keywords args but we should mention positional args first then
# then go with keyword

info('vandana',sal = 60000, age = 22)
```

```python
info(sal = 60000,age = 22,'vandana')
```

```python
# 4. Variable length args
# 1. Variable length positional args
# 2. Variable length keyword args
```

```python
# 1. Variable length positional args--> *args
# when user doesnt have a strict requirement of sending the values to the arguments
# and values varies from one calling to another then use var. len. positional args.
# we need to prefix the argument with *

def test(*n):
  print(n)

test(10)
```

```python
test(10,20,30)

test('python',35,2345345235,'C++')
```

```python
def test(*n):
  print(n)

test('Anshita',20,30)
```

```python
# 2. Variable length keyword args--> **kwargs

def test(**x):
  print(x)

test(num1=10,num2=20,num3=30)
```

```python
test(name='Amol',age=33,sal=23423)
```

```python
# when we want to combine both we can do so
# but remember the rule
# first we need positional n then keyword

def sample(*args,**kwrgs):
  print(args,kwrgs)

sample(10,20,30,num1=200,num2=600)
```
