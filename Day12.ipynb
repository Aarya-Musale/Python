**List comprehension**

```python
#Que: What is List Comprehension?
#Que: Can u explain with the help of an example?
```

```python
[i + 100 for i in range(1,11)]
```

```python
#Syntax: [operation_var for var in sequence]

# with condition if<condition>

#[operation for var in sequence if<condition>]
```

```python
# Need a list of even numbers

[num for num in range(1,21) if num % 2 == 0 ]
```

```python
# Need a list of odd numbers

[num for num in range(1,21) if num % 2 != 0]
```

```python
# Ternary operator
# Syntax
# express_1 if<condition> else express_2
```

```python
gender = ['Male','Female','Female','Male']
# convert male to 0 and female to 1
# if male then 0 else 1

[0 if val == 'Male' else 1 for val in gender]
```

```python
# In list comprehension we cannot use multiple elif conditions
# 1st u can do direct operations
# 2nd u can use if condition
# 3rd u can use ternary operator
```

```python
salary = [24000,12500,45000,65000,11000,10000]
#PS: if sal>25000 than 10% bonus else 20%

[sal + sal * 0.1 if sal> 25000 else sal +sal * .2 for sal in salary]
```

```python
# Nested List Comprehension

# new_list = [[expression for item in inner_iterable] for item in outer_iterable]
```

```python
[val for val in range(1,4)]
```

```python
[[val2 for val2 in range(2)] for val in range(1,4)]
```

```python
[[val for val in range(1,4)] for val2 in range(2)]
```

```python
matrix = [[1,2,3],[4,5,6],[7,8,9]]
odds = [num for row in matrix for num in row if num % 2 != 0]
odds
```
